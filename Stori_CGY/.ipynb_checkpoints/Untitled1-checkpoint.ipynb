```python
df= pd.read_csv('df.csv')
```

```python
df_l = df.copy()
```

```python
df_l["balance"] =df_l["balance"].fillna(df_l["balance"].mean(), inplace = True)
df_l["minimum_payments"] = df_l["minimum_payments"].fillna(df_l["minimum_payments"].mean(), inplace = True)
```

```python
df_l = df[df['activated_date'].notna()]
df_l = df[df['last_payment_date'].notna()]
df_l = df[df['cash_advance'].notna()]
df_l.drop(df_l[df_l.activated_date.isnull()].index, inplace = True)
```

```python
df_l.drop(df_l[df_l.cash_advance.isnull()].index, inplace = True)
df_l.drop(df_l[df_l.last_payment_date.isnull()].index, inplace = True)
```

```python
df_l.drop(df_l[df_l.credit_limit.isnull()].index, inplace = True)
```

```python
y = df_l["fraud"]
#col = [i for i in df_l.columns if i != "fraud" and i!=  "cust_id" and i !="activated_date"and i !="last_payment_date"and i !="last_payment_datec" and i !="activated_datec"]
col = [i for i in df_l.columns if i != "fraud" and i!=  "cust_id" and i !="activated_date"and i !="last_payment_date"and i !="last_payment_datec" and i !="activated_datec" and i !="balance"and i !="oneoff_purchases"and i !="purchases_installments_frequency" ]

X = df_l[col]
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size= .8, random_state = 0 )
```
